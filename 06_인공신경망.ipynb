## 1. 퍼셉트론
 다수의 신호를 입력받아 하나의 신호로 출력


```python
# AND Gate
def and_v1(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2 
    
    if tmp<= theta:
        return 0 
    else : 
        return 1
```

```python
inputs = [(0,0),(1,0),(0,1),(1,1)]

for x1, x2 in inputs:
    y = and_v1(x1, x2)
    print("({x1},{x2}) -> {y}".format(x1=x1, x2=x2, y=y))
```

```python
# OR Gate
def or_v1(v1, v2):
    x=np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.2
    
    tmp = np.sum(w*x) + b
    if tmp <= 0 :
        return 0
    else:
        return 1
```

```python
inputs =[(0,0),(1,0), (0,1), (1,1)]

for x1, x2 in inputs:
    y = or_v1(x1,x2)
    print("({x1},{x2}) -> {y}".format(x1=x1, x2=x2, y=y))
```

```python
# NAND Gate
def nand_v1(v1, v2):
    x=np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b = 0.7
    
    tmp = np.sum(w*x) + b
    if tmp <= 0 :
        return 0
    else:
        return 1
```

```python
inputs =[(0,0),(1,0), (0,1), (1,1)]

for x1, x2 in inputs:
    y = nand_v1(x1,x2)
    print("({x1},{x2}) -> {y}".format(x1=x1, x2=x2, y=y))
```

```python
# XOR Gate
def XOR(x1, x2):
    s1=nand_v1(x1,x2)
    s2=or_v1(x1,x2)
    y=and_v1(s1,s2)
    return y
```

```python
inputs =[(0,0),(1,0), (0,1), (1,1)]

for x1, x2 in inputs:
    y = XOR(x1,x2)
    print("({x1},{x2}) -> {y}".format(x1=x1, x2=x2, y=y))
```
